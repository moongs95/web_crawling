### 네이버 메인화면 크롤링

```python
from urllib.request import urlopen
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
url = 'https://www.naver.com/'
```

```python
driver = webdriver.Chrome()
```

```python
driver.implicitly_wait(3)
```

```python
driver.get(url)
```

```python
page = driver.page_source
```

```python
soup = BeautifulSoup(page, 'html.parser')
soup
```

```python
mid = soup.find_all('ul',{'class':'ContentHeaderView-module__tab_list___BWrWe'})[1]
mid
```

```python
mid_clicktab = mid.find_all('a',{'class':'ContentHeaderView-module__tab___uYoNi'})
mid_clicktab
```

```python
for idx, element in enumerate(mid_clicktab):
    print(idx, element.text)
```
