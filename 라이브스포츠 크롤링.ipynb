```python
win = soup.find_all('span',{'class':"wcl-simpleText_Asp-0 wcl-scores-simpleText-01_pV2Wk"})
a = []
for i in win:
    a.append(i.text)
```

```python
a.count('W')
```

```python
a.count('L')
```

```python
result = {'win':0, 'draw':0, 'lose':0}
```

```python
keys = list(result.keys())
print(keys)
```

```python
for i in win:
    if i.text == 'W':
        result['win'] += 1
    elif i.text == 'D':
        result['draw'] += 1
    elif i.text == 'L':
        result['lose'] += 1
    print(result)
```

```python
print(result)
```

```python
result = {'win':0, 'draw':0, 'lose':0}
for i in range(len(win)):
    if win[i].text == 'W':
        result['win'] += 1
    elif win[i].text == 'D':
        result['draw'] += 1
    elif win[i].text == 'L':
        result['lose'] += 1
    print(result)
```
